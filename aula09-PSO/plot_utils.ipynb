```python
from matplotlib import pyplot as plt
import glob
import os
import shutil
import numpy as np
```

```python
from PIL import Image

class PlotUtils:

    directory = "pso_plots"
    filename = "pso.gif"

    @staticmethod
    def start_plot():
        # Remove o diretório existente e cria um novo para armazenar as iterações
        if os.path.exists(PlotUtils.directory):
            shutil.rmtree(PlotUtils.directory)  # Remove o diretório se existir
        os.makedirs(PlotUtils.directory)    # Cria um novo diretório 

    @staticmethod
    def plot_particle(particle):
        # Plota a posição de uma partícula em um gráfico
        plt.scatter(particle.position[0], particle.position[1])

    @staticmethod
    def plot_iteration(i):
        plt.title(f"PSO {i}")
        plt.xlim(-1500, 1500) # Limites do eixo x
        plt.ylim(-1500, 1500) # Limites do eixo y
        plt.xlabel('x[0]')
        plt.ylabel('x[1]')
        iteration = str(i).zfill(5)
        plt.savefig(f"pso_plots/iteration_{iteration}.png", facecolor = "white", dpi = 75)
        plt.close()

    @staticmethod
    def save():
        images = [Image.open(f) for f in sorted(glob.glob(PlotUtils.directory+"/*"))]
        
        if not images: 
            print("Nenhuma imagem encontrada para criar o GIF.")
            return

        img = images[0]  # Usa a primeira imagem como base
        img.save(fp=PlotUtils.filename, format='GIF', append_images=images, save_all=True, duration=200, loop=0)
        
        print(f"GIF salvo como {PlotUtils.filename}")

        if os.path.exists(PlotUtils.directory):
            shutil.rmtree(PlotUtils.directory)


class Particle:
    def __init__(self, position):
        self.position = position
        
PlotUtils.start_plot()

for i in range(10):  # 10 iterações
        plt.figure()
        
        # Criando partículas com posições aleatórias
        particles = [Particle(np.random.uniform(-1500, 1500, size=2)) for _ in range(5)]  # 5 partículas
        for particle in particles:
            PlotUtils.plot_particle(particle)  # Plota as partículas
        PlotUtils.plot_iteration(i)  


PlotUtils.save()  
```
